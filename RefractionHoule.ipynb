# Modélisation de la réfraction de la houle par tracé de rayons

On s'intéresse à la réfraction de la houle par une bathymétrie. En supposant que les longueurs d'onde des ondes de surface sont petites devant l'échelle de variation de la profondeur, on effectue un tracé de rayons.

## Bathymétrie

On définit la profondeur de par l'expression
$$h(x,y) =h_{large} -
  \left[1+ \mu \, \sin\left( {2\, \pi\, x \over D}\right)\right]
  \,  \left[ \lambda \, (y-y_{plat}) + \sqrt{\lambda^2 \, (y-y_{plat})^2 + 4} \right]/2\,$$
telle que $h\sim h_{large}$ pour $y\ll y_{plat}$ et où $\lambda$ est la pente moyenne en $y$ du fond, modulée en $x$ avec le facteur $\mu$. La cote du fond est alors $Z_f(x,y) = - h(x,y)$ (iso-profondeurs en pointillés sur la figure 1). La ligne $Z_f(x,y)=0$ (en trait plein) définit le trait de côte.

## Relation de dispersion des ondes de surface

Lorsque la profondeur $h$ est constante, les ondes de surface planes et monochromatiques de la forme
$$u(x,y,t) = A \, \cos(\underline k \cdot \underline x - \omega \, t + \phi)\;,$$
 de vecteur d'onde   $\underline k = k_{x} \, \underline e_x + k_{y} \, \underline e_y$, sont solutions des équations d'Euler irrotationnelles si et seulement la relation de dispersion suivante est vérifiée :
$$ \omega  = \Omega(k_x, k_y) =\Omega(\underline k)=  \Omega_{houle} (k) = \sqrt{g\, k \tanh(k\, h)} \;,$$
où $g$ est la gravité et $k= \| \underline k\| = \sqrt{k_x^2 + k_y^2}$.
On en déduit les expressions suivantes des vitesses de phase $\underline c_\varphi$ et de groupe $\underline c_g$ :
\begin{eqnarray}
\underline c_\varphi(\underline k)  &= {\Omega(\underline k)\over k} \,\underline e_k=  c_\varphi(k) \, \underline e_k
 & \hbox{avec} \quad
c_\varphi(k) = \sqrt{{g\over k} \, \tanh(k\, h)} \;, \cr
\underline c_g(\underline k) &= {\mkern10mu\underline{\quad}\mkern-28mu{\hbox{grad}}_{\underline k}} \Omega(\underline k) =  c_g(k) \, \underline e_k
 & \hbox{avec} \quad
c_g(k)  ={1\over 2} \,  c_\varphi(k) \, \left[ 1 + {2\, k \, h \over \sinh(2\, k\, h)} \right]  \;,
\end{eqnarray}
où ${\mkern10mu\underline{\quad}\mkern-28mu{\hbox{grad}}_{\underline k}}  = {\partial \over \partial k_x} \, \underline e_x + {\partial \over \partial k_y} \, \underline e_y $ est le gradient par rapport aux variable $k_x$ et $k_y$.

## Relations de dispersion locale pour une bathymétrie non homogène

Lorsque l'échelle de variation de la profondeur $h(x,y)$ est grande devant la longueur d'onde $L_0 = 2\, \pi / k_0$, avec $k_0 = \| \underline k_0\|$, on considère des paquets d'onde de la forme $u(x,y,t) = A(x,y)\, \cos[\varphi(x,y,t)]$ et on définit les champs suivants :
$$  k_x(x,y,t) = {\partial \varphi\over \partial x} (x,y,t)
  \;, \quad
  k_y(x,y,t) = {\partial \varphi\over \partial y}  (x,y,t)
   \quad \hbox{et} \quad
  \omega(x,y,t) =- {\partial \varphi\over \partial t} (x,y,t)\;.$$
L'approximation WKB (non détaillée ici) permet d'affirmer que la relation de dispersion locale suivante est vérifiée :
 $$ \omega = \Omega(k_x, k_y, x, y)
  \quad \Longleftrightarrow \quad
{\partial \varphi\over \partial t} + \Omega\left( {\partial \varphi\over \partial x} , {\partial \varphi\over \partial y}  , x, y\right) = 0\;,$$
où $\Omega(\underline k, \underline x)$ est obtenu en tenant compte de la variation de $h(\underline x)$ dans la relation de dispersion homogène.  

En prenant le gradient en $\underline x$ de cette équation de l'Eikonale, qui s'écrit  $ {\partial \varphi\over \partial t}  + \Omega(  {\mkern10mu\underline{\quad}\mkern-28mu{\hbox{grad}}}\,\varphi, \underline x) =0$, et en utilisant la relation ${\partial k_x\over \partial y}={\partial k_y\over \partial x}={\partial^2 \varphi\over \partial x \, \partial y}$, on obtient une équation de transport pour $\underline k$ qui s'écrit :
$$ {\partial \underline k \over \partial t}  + \underline c_g(\underline k,\underline x) \cdot  {\mkern10mu\underline{\quad}\mkern-28mu{\hbox{grad}}}\,   \underline k = - {\mkern10mu\underline{\quad}\mkern-28mu{\hbox{grad}}}_{\underline x}\,\Omega(\underline k,  \underline x)\,$$
où $\underline c_g \cdot {\mkern10mu\underline{\quad}\mkern-28mu{\hbox{grad}}} = c_{gx} \, {\partial \over \partial x} + c_{gy} \, {\partial \over \partial y}$ est l'opérateur de dérivation selon la vitesse $\underline c_g$ avec $c_{gx} = {\partial \Omega \over \partial k_x}$ et $c_{gy} = {\partial \Omega \over \partial k_y}$, et ${\mkern10mu\underline{\quad}\mkern-28mu{\hbox{grad}}_{\underline x}} \Omega(\underline k, \underline x)= {\partial \Omega \over \partial x}\, \underline e_x + {\partial \Omega \over \partial y}\, \underline e_y$.

## Onde plane incidente

On suppose qu'une onde plane arrive d'une zone où la profondeur $h_0$ est constante,en faisant un angle $\theta$ avec l'axe $Oy$. L'élévation de la surface de la mer s'écrit alors $u(x,y,t) = A_0 \, \cos(\underline k_0 \cdot \underline x - \omega_0 \, t + \phi_0)$ avec $\underline x = x \, \underline e_x + y \, \underline e_y$ et $\underline k_0 = k_{0x} \, \underline e_x + k_{0y} \, \underline e_y$ où $\underline e_x$ et $\underline e_y$ sont les vecteurs de base, la relation de dispersion $\omega_0 =\sqrt{g\, k_0 \tanh(k\, h_0)}$ étant vérifiée. On suppose connue cette expression sur un segment de droite $\cal L$ perpendiculaire au vecteur d'onde $\underline k_0$ et situé suffisamment loin de la côte (ligne verte en gras sur la figure 1).

## Tracé de rayons

La théorie des caractéristique (non détaillée ici), permet de résoudre l'équation de transport du vecteur d'onde $\underline k (\underline x, t)$ en calculant une famille de trajectoires $\underline  x_{\cal C}(t)=x_{\cal C}(t) \, \underline e_x + y_{\cal C}(t) \, \underline e_y$ de l'espace transportant le vecteur $\underline k_{\cal C}(t) = k_{x_{\cal C}}(t) \, \underline e_x + k_{y_{\cal C}}(t) \, \underline e_y$, solutions du  système d'équations différentielles ordinaires (EDO) suivant :
$$ {d \underline x_{\cal C} \over dt }  ={\mkern10mu\underline{\quad}\mkern-28mu{\hbox{grad}}_{\underline k}}\,\Omega(\underline k_{\cal C}, \underline x_{\cal C})  = \underline c_g(\underline k_{\cal C}, \underline x_{\cal C})
\quad \hbox{et} \quad
 {d \underline k_{\cal C} \over dt }  = - {\mkern10mu\underline{\quad}\mkern-28mu{\hbox{grad}}_{\underline x}}\, \Omega(\underline k_{\cal C}, \underline x_{\cal C})\;.$$
 En utilisant l'expression de la relation de dispersion des ondes de surface, on obtient le sytème suivant d'équations différentielles ordinaires (EDO) :
\begin{equation}
\begin{matrix}
{d x \over dt }  & =
 {1\over 2} \,  {\sqrt{g\, k  \, \tanh(k\, h)}\over k}\,
 \left[1 + {2\, k \, h \over \sinh(2\, k\, h)} \right]\,{k_x \over k}\;,
 & \quad
 {d k_x \over dt }  & =
 -k {\sqrt{ g\, k  \, \tanh(k\, h)} \over \sinh(2\,  k\, h)} \, {\partial h \over \partial x } (x,y)\;,
 \cr
{dy \over dt }  & =
 {1\over 2} \,   {\sqrt{g\, k  \, \tanh(k\, h)}\over k}\,
 \left[1 + {2\, k \, h \over \sinh(2\, k\, h)} \right]\,{k_y \over k}\;,
 & \quad
 {d k_y \over dt } & =
 -k {\sqrt{ g\, k  \, \tanh(k\, h)} \over \sinh(2\,  k\, h)} \, {\partial h \over \partial y } (x,y)\;.
\end{matrix}
\end{equation}
On considère alors les conditions initiales de ces trajectoire pour tout point $\underline x_0 = x_0 \, \underline e_x + y_0 \, \underline e_y$ du segment de droite $\cal L$ en posant
$$  x(0) = x_0 \;, \quad y(0) = y_0 \;, \quad k_x(0) = k_{x0} \;, \quad  k_y(0) = k_{y0} \;. $$
 La forme hamiltionienne de ce système entraine que $\omega(\underline x, t)$ est constante le long des trajectoires, et donc égale à $\omega_0$ pour tout pour $\underline x$. On détermine le vecteur d'onde $\underline k(\underline x)$ en tout point $\underline x$ par lequel passe une trajectoire $\underline x_{\cal C}(t)$ au temps $t$ en écrivant
 $\underline k(\underline x) =\underline k_{\cal C}(t)$. De même, la phase $\varphi_0(\underline x)$ du paquet d'onde au temps $t=0$ est
  $\varphi_0(\underline x) = \varphi_0[\underline x_{\cal C}(t)]$ avec
\begin{eqnarray}
 \varphi_o[\underline x_{\cal C}(t)]  & =&  \underline k_0\cdot \underline x_0 + \phi_0
 + \int_0^t c_{g_{\cal C}}(t') \, k_{\cal C}(t')\, dt' \;,
 \cr
 \hbox{avec} \quad
 k_{\cal C}(t) = \|\underline k_{\cal C}(t)\|
\quad \hbox{et} &&
  c_{g_{\cal C}} = {1\over 2} \,  {\sqrt{g\, k_{\cal C}  \, \tanh[k_{\cal C}\, h(\underline x_{\cal C})]}\over k_{\cal C}}\,
 \left[1 + {2\, k_{\cal C} \, h(\underline x_{\cal C})  \over \sinh[2\, k_{\cal C}\, h(\underline x_{\cal C})]} \right]\;.
\end{eqnarray}
Comme le segment de droite $\cal L$ est perpendiculaire au vecteur d'onde $\underline k_0$, on peut choisir $\phi_0 = - \underline k_0 \cdot \underline x_0$ pour n'importe quel point $\underline x_0$ de $\cal L$, afin que la phase à $t=0$ soit nulle sur ce segment de droite.

## Conservation de l'énergie

L'énergie par unité de surface d'une houle monochromatique de la forme $u(x,y,t) = A \, \cos(k_x \, x+ k_y\, y - \omega\, t+ \phi)$ est $E={1\over 2} \, \rho\, g\, A^2$, où $\rho$ est la masse volumique de l'eau. L'amplitude $A$ d'un paquet d'onde de la forme $u(\underline x,t) = A(\underline x,t)\, \cos[\varphi(\underline x,t)]$ émis par l'onde plane monochromatique du segment de droite $\cal L$ ne dépend pas du temps et vérifie
$$
{\rm div} \left\{ A^2(\underline x)\,  \underline c_g[\underline k(\underline x)] \right\}
  = {\partial \over \partial x}\left[ A^2(x,y) \,  c_{gx}  \right] +{\partial \over \partial x}\left[  A^2(x,y)\, c_{gy} \right] = 0\;.$$
En intégrant cette relation sur un domaine $\cal D$ compris entre deux trajectoires $\cal C_1$ et $\cal C_2$ et délimité par deux segments de droites $P_1P_2$ et $Q_1Q_2$ (figure 3), on montre que
  $$ |P_1P_2| \, A_P^2 \;\left( \underline c_{g_P} \cdot \underline e_P\right)
=  |Q_1 Q_2| \, A_Q^2 \;\left(  \underline c_{g_Q} \cdot \underline e_Q\right)\;, $$
où $(A^2_P\, \underline c_{gP})$ et $(A^2_Q\,  \underline c_{gQ})$ sont respectivement les valeurs moyennes de $A^2\, \underline c_{g}$ sur les segments $P_1P_2$ et $Q_1Q_2$, et $\underline e_P$ et $\underline e_Q$ les vecteurs unitaires perpendiculaires à ces segments dans la direction des trajectoires.
La hauteur $H=2\,A$ de la houle augmente donc si les trajectoires des rayons se rapprochent ou si la vitesse de groupe diminue. Les vagues déferlent si la hauteur $H$ dépasse la valeur critique $H_c$ donnée par le critère de déferlement de Miche :
$$H_c = 0,14\, {2\, \pi \over k} \, \tanh(k\, h) \;. $$
Dans ce cas, on annule numériquement la hauteur $H$, ce qui permet de visualiser la ligne de déferlement lors du tracé des iso-contours du champ $H(x,t)$, comme on peut le voir sur la figure 1.

```python
# Sous-progammes a compiler

# Import de librairies
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation


def inifig(fx,fy,xaxe=0,yaxe=0,tit='Titre',xlab='x',ylab='y'):
    # initialisation d'une figure
    fig=plt.figure(figsize=(fx,fy))
    plt.axvline(x=xaxe)
    plt.axhline(y=yaxe)
    plt.xticks(fontsize=12)
    plt.yticks(fontsize=12)
    plt.xlabel(xlab,fontsize=16)
    plt.ylabel(ylab,fontsize=16)
    plt.title(tit,fontsize=16)
    plt.minorticks_on()
    plt.grid(which='minor',color='grey', axis='y', linestyle=':', linewidth=1)
    plt.grid(color='black', axis='y', linestyle='-',linewidth=1)
    plt.grid(color='black', axis='x', linestyle='-',linewidth=1)
    return fig


def bathy(x,y):
    # bathymetrie h(x,y)
    kD=2*np.pi/D
    yt=la*(y-yplat); sq=np.sqrt(yt**2+4)
    h=hlarge-(1+mu*np.sin(kD*x))*.5*(yt+sq)
    hx=-mu*kD*np.cos(kD*x)*.5*(yt+sq)
    hy=-(1+mu*np.sin(kD*x))*.5*la*(1+yt/sq)
    return h,hx,hy

def visubathy(x,y):
  # Trace de Zf=-h(x,y)
  X, Y = np.meshgrid(x, y)
  h,Hx,Hy=bathy(X,Y)
  Zf=-h
  # Isolignes
  fig=inifig(fx,fy,tit=r'Tracé de rayons')
  contours = plt.contour(X, Y, Zf, levels=10, colors='black')  # contour lines
  plt.clabel(contours, inline=True, fontsize=16, fmt="%.0f")   # label the lines
  return fig

def visubathy3D(x,y):
  # Trace de  Zf=-h(x,y) en 3D
  X, Y = np.meshgrid(x, y)
  h,Hx,Hy=bathy(X,Y)
  Zf=-h
  # Color map
  # plot 3D
  fig = plt.figure(figsize=(9, 9))
  ax = fig.add_subplot(111, projection='3d')
  # clip Zf
  Zf=np.where(Zf<0, Zf,0)
  Nyh=0;
  Xtron=np.zeros((Nyh,Nx));
  Ytron=np.zeros((Nyh,Nx));
  Zftron=np.zeros((Nyh,Nx));
  Xtron=X[Nyh:Ny,:]; Ytron=Y[Nyh:Ny,:]; Zftron=Zf[Nyh:Ny,:]
  ax.plot_wireframe(Xtron, Ytron, Zftron, color='k', linewidth=0.5)
  ax.view_init(elev=20, azim=-40)  # tilt elev, rotate azim around z-axis
  ax.set_xlim(xmin, xmax)
  ax.set_ylim(ymin, ymax)
  #ax.set_zlim(-15, 0)
  #fig.colorbar(surf, ax=ax, shrink=.6, label=r'$Z_f(x, y)$')
  ax.set_xlabel('x')
  ax.set_ylabel('y')
  ax.set_zlabel(r'$Z_f$')
  #ax.set_title(r'Batymétrie $Z_f(x, y)$')
  fig.savefig('visubathy3D.pdf')
  return fig

def bathyclip(x,y,la,mu,D):
    # bathymetrie h(x,y) avec clip en hmin
    h,hx,hy = bathy(x,y)
    h=np.where(h >0, h, hmin)
    hx=np.where(h >0, hx, 0)
    hy=np.where(h >0, hy, 0)
    return h,hx,hy

def houle(k,h):
  # vitesses de phase et de groupe
  kh=k*h
  om=np.sqrt(g*k*np.tanh(kh))
  cphi=om/k;
  cg=.5*cphi*(1+2*kh/np.sinh(2*kh))
  return om, cphi, cg

def visuhoule(h,kmax):
  # visualisation de omega, cphi et cg
  k=np.linspace(.001,kmax,500)
  om, cphi,cg=houle(k,h)
  xla=r'$k$';  yla=r'$\omega$, $c_\varphi$, $c_g$';
  ti=r'Relation de dispersion $\omega= \sqrt{g\, k\, \tanh(k\,h)}$'
  fig=inifig(8.6,5,xlab=xla,ylab=yla,tit=ti)
  plt.plot(k,om,color="green",linewidth=4)
  plt.plot(k,cphi,color="blue",linewidth=4)
  plt.plot(k,cg,color="red",linewidth=4)
  fig.savefig('visuhoule.pdf')

def rayon(x,y,kx,ky,dt,Nt):
  # trace d'un rayon unique
  # initialisation des tableaux 1D
  xr=np.zeros(Nt); yr=np.zeros(Nt);
  kxr=np.zeros(Nt); kyr=np.zeros(Nt);
  cgxr=np.zeros(Nt); cgyr=np.zeros(Nt);
  phir=np.zeros(Nt); hr=np.zeros(Nt);
  phi=kx*x+ky*y+phi0;
  # boucle sur le temps
  for i in range(Nt):
    k=np.sqrt(kx**2+ky**2)
    h,hx,hy=bathyclip(x,y,la,mu,D)
    om,cphi,cg=houle(k,h)
    cgx=cg*kx/k; cgy=cg*ky/k;
    # stockage du rayon dans les tableaux
    xr[i]=x; yr[i]=y;
    kxr[i]=kx; kyr[i]=ky;
    cgxr[i]=cgx; cgyr[i]=cgy;
    phir[i]=phi; hr[i]=h;
    # avance seulement pour h>hmin
    if h>hmin:
      # avance en x,y
      x=x+cgx*dt; y=y+cgy*dt;
      # avance en phi
      phi=phi+cg*k*dt
      # avance en kx, ky
      komsinh=k*om/np.sinh(2*k*h)
      Fkx=-komsinh*hx; Fky=-komsinh*hy;
      kx=kx+Fkx*dt; ky=ky+Fky*dt

  return xr,yr,kxr,kyr,cgxr,cgyr,phir,hr

def rayons(th):
  # calcul de tous les rayons

  # initialisation des tableaux 2D
  xl=np.zeros((Ns+1,Nt)); yl=np.zeros((Ns+1,Nt));
  kxl=np.zeros((Ns+1,Nt)); kyl=np.zeros((Ns+1,Nt));
  cgxl=np.zeros((Ns+1,Nt)); cgyl=np.zeros((Ns+1,Nt));
  phil=np.zeros((Ns+1,Nt)); hl=np.zeros((Ns+1,Nt));

  # boucle sur les points de depart de L
  dsx=ds*np.cos(th); dsy=-ds*np.sin(th)
  for n in range(Ns+1):
    xs=xd+n*dsx; ys=yd+n*dsy;
    # calcul des rayons
    xr,yr,kxr,kyr,cgxr,cgyr,phir,hr=rayon(xs,ys,kx0,ky0,dt,Nt)
    xl[n,:]=xr; yl[n,:]=yr;
    kxl[n,:]=kxr; kyl[n,:]=kyr;
    cgxl[n,:]=cgxr; cgyl[n,:]=cgyr;
    phil[n,:]=phir; hl[n,:]=hr;

  return xl,yl,kxl,kyl,cgxl,cgyl,phil,hl

def visurayons():
  # Visualisation des rayons

  # plots de la bathymetrie et de la ligne d'emission L
  fig=visubathy(x,y);
  plt.plot([xd,xd+2*D],[yd,yd-2*D*np.sin(th)],color="green",linewidth=3)

  # Calcul de tous les rayons
  xl,yl,kxl,kyl,cgxl,cgyl,phil,hl=rayons(th)

  # Energie
  #########
  # initialisation des champs intermediaires
  xm=np.zeros((Ns,Nt));ym=np.zeros((Ns,Nt))
  cgxm=np.zeros((Ns,Nt));cgym=np.zeros((Ns,Nt))
  edlxm=np.zeros((Ns,Nt));edlym=np.zeros((Ns,Nt))
  Em=np.zeros((Ns,Nt)); hm=np.zeros((Ns,Nt))
  km=np.zeros((Ns,Nt))

  # calculs aux points intermediaires
  xm=.5*(xl[0:Ns,:]+xl[1:Ns+1,:]);
  ym=.5*(yl[0:Ns,:]+yl[1:Ns+1,:]);
  hm=.5*(hl[0:Ns,:]+hl[1:Ns+1,:]);
  cgxm=.5*(cgxl[0:Ns,:]+cgxl[1:Ns+1,:]);
  cgym=.5*(cgyl[0:Ns,:]+cgyl[1:Ns+1,:]);
  kl=np.sqrt(kxl**2+kyl**2);
  km=.5*(kl[0:Ns,:]+kl[1:Ns+1,:]);
  dxm=xl[1:Ns+1,:]-xl[0:Ns,:];
  dym=yl[1:Ns+1,:]-yl[0:Ns,:];
  edlxm=-dym; edlym=dxm;

  # Initialisation de la hauteur de houle
  Em[:,0]=np.zeros(Ns)+E0
  for i in range(1,Nt):
    cgdelim1=cgxm[:,i-1]*edlxm[:,i-1]+cgym[:,i-1]*edlym[:,i-1]
    cgdeli=cgxm[:,i]*edlxm[:,i]+cgym[:,i]*edlym[:,i]
    Em[:,i]=Em[:,i-1]*cgdelim1/cgdeli

  # visualisation hauteur de la houle Hm
  Em=np.where(Em>0, Em, 0)
  Hm=2*np.sqrt(Em);
  Hmc=.14*(2*pi)/km*np.tanh(km*hm) # critere de Miche
  Hm=np.where(Hm < Hmc, Hm, 0)
  contours = plt.contour(xm,ym,Hm,levels=30,colors='brown')  # contour lines
  plt.clabel(contours, inline=True, fontsize=16, fmt="%.1f")   # label the lines

  # Phase
  #########
  contours = plt.contour(xl,yl,phil/(2*pi),levels=20,colors='green',linewidths=2)  # contour lines
  plt.clabel(contours, inline=True, fontsize=16, fmt="%.1f")   # label the lines

  # Rayons
  #########
  Nsr=Ns//Nr
  for n in range(Ns+1)[::Nsr]:
    # trace des rayons
    xr=xl[n,:]; yr=yl[n,:];
    plt.plot(xr,yr,color="blue",linewidth=1)
    # trace des vitesses de groupe
    Ntq=Nt//10;
    xq=xl[n,::Ntq]; yq=yl[n,::Ntq]; kxq=kxl[n,::Ntq]; kyq=kyl[n,::Ntq];
    cgxq=cgxl[n,::Ntq]; cgyq=cgyl[n,::Ntq]
    plt.quiver(xq,yq,cgxq,cgyq,color='red',scale=100)
    plt.axis('equal')

  ##############
  plt.xlim(xmin,xmax)
  plt.ylim(ymin,ymax)
  fig.savefig('visurayons.pdf')
  fig.savefig('visurayons.jpg')

  return xl,yl,kxl,kyl,cgxl,cgyl,phil,hl,xm,ym,Hm
```

```python
# Redéfinition de sous-programmes
def bathy(x,y):
    # bathymetrie h(x,y) déclée de D/2
    kD=2*np.pi/D
    yt=la*(y-yplat); sq=np.sqrt(yt**2+4)
    h=hlarge-(1+mu*np.sin(kD*(x-D/2)))*.5*(yt+sq)
    hx=-mu*kD*np.cos(kD*(x-D/2))*.5*(yt+sq)
    hy=-(1+mu*np.sin(kD*(x-D/2)))*.5*la*(1+yt/sq)
    return h,hx,hy
```
